##### Q3 > Answer the Following

a. What is the average length of stay in hospital of a female patient who 
was readmitted in less than 30 days?

-> Here we have given a series of conditions to take only selected data into the "clean_data" column here it takes the data which has people which were READMITTED WITHIN 30 DAYS ;  GENDER where ONLY FEMALE and LENGTH OF STAY which was LESSER THAN 30 days

```python
clean_data = data[(data["readmitted"] == "<30") & (data["gender"] == "Female") & (data["length_of_stay"] < 30)] # cleaning of data

avg_stay_length = clean_data["length_of_stay"].mean()

print("Average of Total Days Female staying -> who was readmitted within 30 days", avg_stay_length)
```

##### b. Which age group has the highest risk of being readmitted within 30 days?

-> Here we clean the data to keep only the people which were readmitted to the 30 days then store this value into the "clean_data" column further more we take "age" column from the clean data and count the number of people which fall under this category and then plot a line plot

```python
clean_data = data[(data['readmitted'] == "<30")]
age_cnt = clean_data['age'].value_counts()
age_cnt.plot()

print("\n\n | Maximum people with age [70-80] years old were readmitted within 30 days |")
```

##### c. Which age groups have more than 4000 cases of being readmitted? 

-> Here we can see that We have cleaned the data to keep only the people which were readmitted we store that value into column "clean_data" Then we take "age" column from the cleaned data encounter how many people were above 4000 range and store it to "age_cnt" variable

```python
clean_data =  data[(data['readmitted'] != "No")]
age_cnt = clean_data['age'].value_counts()

above_4k = age_cnt[age_cnt > 4000]
print(above_4k)
print("\n\n | Maximum people with age [40-80] years old were readmitted more than 4000 times |")
plt.plot(above_4k)
```

##### d. Which are the top-three race categories according to the number of readmission cases?

-> Here we can see that we have cleaned the data by only keeping the people who were readmitted and then we stored it in a variable named "clean_data" then we process "clean_data" to keep only the top 3 values having the most re-admission cases

```python
clean_data = data[data["readmitted"] != "No"]
top_three = clean_data["race"].value_counts().head(3)

print(" \n | TOP 3 RACE AS PER RE-ADMISSION CASES ARE | \n\n")

print(top_three)
```
